```python
def load_data(response):
    # Parse the response text using BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find the script tag by id
    script_tag = soup.find('script', id='index-data')

    if script_tag is not None:
        # Extract the JSON content from the script tag
        json_content = script_tag.string

        # Parse the JSON content using Python's json module
        data = json.loads(json_content)

        return data
    else:
        return None
    


def append_values(data):
    values_list = []
    for item in data['grid']['items']:
        values_list.append([
            item.get('rawPrice'),
            item.get('price'),
            item.get('priceWithFees')
        ])
    return values_list


```

```python


# Function to print all keys recursively
def print_keys(data, parent_key=''):
    if isinstance(data, dict):
        for key, value in data.items():
            new_key = f"{parent_key}.{key}" if parent_key else key
            print_keys(value, new_key)
    elif isinstance(data, list):
        for i, item in enumerate(data):
            new_key = f"{parent_key}[{i}]"
            print_keys(item, new_key)
    else:
        print(parent_key)

# Print all keys
print_keys(data)
```
